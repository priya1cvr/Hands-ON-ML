In Wards Criterior we look at each pair of cluster and see how much the variance would increase if we join them together.<br>
After joining ,if they are very far apart ,variance would be large.So, to avoid that we join where variance is less.


```python
from __future__ import print_function, division
from future.utils import iteritems
from builtins import range, input

import numpy as np
import matplotlib.pyplot as plt

from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
D = 2 # so we can visualize it more easily
s = 4 # separation so we can control how far apart the means are
mu1 = np.array([0, 0])
mu2 = np.array([s, s])
mu3 = np.array([0, s])
```

```python
mu1,mu2,mu3
```

```python
N = 900 # number of samples
X = np.zeros((N, D))
X[:300, :] = np.random.randn(300, D) + mu1
X[300:600, :] = np.random.randn(300, D) + mu2
X[600:, :] = np.random.randn(300, D) + mu3
```

```python
X
```

```python
Z = linkage(X, 'ward')
print("Z.shape:", Z.shape)
```

```python
plt.title("Ward")
dendrogram(Z)
plt.show()
```

```python
Z = linkage(X,'single')
plt.title("Single")
dendrogram(Z)
plt.show()
```

```python
Z = linkage(X,'complete')
plt.title("Complete")
dendrogram(Z)
plt.show()
```
